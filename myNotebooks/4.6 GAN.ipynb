# 4.6 GAN

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
BATCH_SIZE = 64
LR_G = 0.0001
LR_D = 0.0001
N_IDEAS = 5  #灵感，即输入数据大小
ART_COMPONENTS = 15
#横坐标线段从-1到1，共15个（ART_COMPONENTS）点
PAINT_POINTS = np.vstack(np.linspace(-1,1,ART_COMPONENTS) for i in range(BATCH_SIZE))
print(PAINT_POINTS.size)

# show our beautiful painting range
plt.plot(PAINT_POINTS[0], 2 * np.power(PAINT_POINTS[0], 2) + 1, c='#74BCFF', lw=3, label='upper bound')
plt.plot(PAINT_POINTS[0], 1 * np.power(PAINT_POINTS[0], 2) + 0, c='#FF9359', lw=3, label='lower bound')
plt.legend(loc='upper right')
plt.show()
```

```python
def artist_works():
    #随机产生曲线
    a = np.random.uniform(1,2,size=BATCH_SIZE)[:,np.newaxis] #区间为1~2的随机数
    paintings = a*np.power(PAINT_POINTS,2)+(a-1)
    paintings = torch.from_numpy(paintings).float()
    return paintings
```

### 定义网络

```python
#生成器
G = nn.Sequential(
    nn.Linear(N_IDEAS,128),
    nn.ReLU(),
    nn.Linear(128,ART_COMPONENTS),
)
#判别器
D = nn.Sequential(
    nn.Linear(ART_COMPONENTS,128),
    nn.ReLU(),
    nn.Linear(128,1),
    nn.Sigmoid(), #将判别结果转换为百分比的形式
)

opt_D = torch.optim.Adam(D.parameters(),lr=LR_D)
opt_G = torch.optim.Adam(G.parameters(),lr=LR_G)
```

### 训练

```python
for step in range(10000):
    artist_paintings = artist_works()  #真实数据
    G_ideas = torch.randn(BATCH_SIZE,N_IDEAS)
    G_paintings = G(G_ideas)
    
    prob_artist0 = D(artist_paintings) #判断真实的画
    prob_artist1 = D(G_paintings)      #判断假画家（生成器）的画
    
    #D——尽量增加prob_artist0，减少prob_artist1 
    D_loss = -torch.mean(torch.log(prob_artist0)+torch.log(1-prob_artist1))
    #G——增加prob_artist1 
    G_loss = torch.mean(torch.log(1-prob_artist1))
    
    opt_D.zero_grad()
    D_loss.backward(retain_graph=True) #反向传播之后保留计算梯度，以便G更新时使用
    opt_D.step()
    
    opt_G.zero_grad()
    G_loss.backward()  #没有设置retain_graph，执行完后中间变量释放
    opt_G.step()
    
    if step % 1000 == 0:  # plotting
        plt.cla()
        plt.plot(PAINT_POINTS[0], G_paintings.data.numpy()[0], c='#4AD631', lw=3, label='Generated painting',)
        plt.plot(PAINT_POINTS[0], 2 * np.power(PAINT_POINTS[0], 2) + 1, c='#74BCFF', lw=3, label='upper bound')
        plt.plot(PAINT_POINTS[0], 1 * np.power(PAINT_POINTS[0], 2) + 0, c='#FF9359', lw=3, label='lower bound')
        plt.text(-.5, 2.3, 'D accuracy=%.2f (0.5 for D to converge)' % prob_artist0.data.numpy().mean(), fontdict={'size': 15})
        plt.text(-.5, 2, 'D score= %.2f (-1.38 for G to converge)' % -D_loss.data.numpy(), fontdict={'size': 15})
        plt.ylim((0, 3));plt.legend(loc='upper right', fontsize=12);plt.draw();plt.pause(0.01)
        plt.show()
```
