# 3.1 回归

```python
import torch
import torch.nn.functional as fun
import matplotlib.pyplot as plt
```

### 1. 生成假数据 
- torch.squeeze()：对数据维度进行压缩，去掉维数为1的维度  
- torch.unsqueeze()：给指定位置加上维数为一的维度

```python
x = torch.unsqueeze(torch.linspace(-1,1,100),dim=1) # shape=(100,1)
y = x.pow(2) + 0.2*torch.rand(x.size())  #含有噪声的y，shape=(100,1)

%matplotlib inline
plt.scatter(x.numpy(),y.numpy()) #散点图
plt.show()
```

### 2. 定义neural network

```python
class Net(torch.nn.Module): #继承torch.nn.Module模块
    def __init__(self,n_feature,n_hidden,n_output):
        super(Net,self).__init__() #继承
        #定义网络
        self.hidden = torch.nn.Linear(n_feature,n_hidden)
        self.predict = torch.nn.Linear(n_hidden,n_output)
    
    def forward(self,x): #前向传播
        x = torch.relu(self.hidden(x))
        x = self.predict(x) #回归最后一层一般不用激励函数
        return x
    
net = Net(1,10,1) #一个输入对应一个输出，隐藏层有10个神经元
print(net)
optimizer = torch.optim.SGD(net.parameters(),lr=0.5)
loss_func = torch.nn.MSELoss() #均方误差
```

### 3. 训练


```python
#%matplotlib notebook
%matplotlib inline
plt.ion() #实时显示（遇到plt.show()的时候进程不阻塞）
plt.show()

for t in range(100):
    prediction = net(x)
    
    loss = loss_func(prediction,y)
    
    optimizer.zero_grad() #梯度降为0
    loss.backward()       #反向传递，计算梯度
    optimizer.step()      #优化参数
    
    #画图
    if t % 10 == 0:
        plt.cla()
        plt.scatter(x.numpy(), y.numpy())
        plt.plot(x.numpy(), prediction.data.numpy(), 'r-', lw=5)
        plt.text(0.5, 0, 'Loss=%.4f' % loss.data.numpy(), fontdict={'size': 20, 'color':  'red'})
        plt.show()
        plt.pause(0.1)

plt.ioff()
```
