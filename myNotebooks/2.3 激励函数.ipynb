### 2. 看看激励函数长啥样

```python
import torch
import torch.nn.functional as fun
import matplotlib.pyplot as plt
```

- **生成自变量x的假数据**

```python
x = torch.linspace(-5,5,20)
x_np = x.numpy() #画图时要转换成numpy
```

- **生成不同函数对应的函数值y**

```python
y_relu = torch.relu(x).numpy()
y_sigmoid = torch.sigmoid(x).numpy()
y_tanh = torch.tanh(x).numpy()
y_softplus = fun.softplus(x).numpy() #torch没有softplus
#y_softmax= torch.softmax(x).numpy()
```

- **画图**

```python
%matplotlib inline
plt.figure(1,figsize=(10,8))
plt.subplot(221)
plt.plot(x_np,y_relu,label='relu')
plt.ylim((-1,5))
plt.legend(loc='best')

plt.subplot(222)
plt.plot(x_np,y_sigmoid,label='sigmoid')
plt.ylim((-0.2,1.2))
plt.legend(loc='best')

plt.subplot(223)
plt.plot(x_np,y_tanh,label='tanh')
plt.ylim((-1.2,1.2))
plt.legend(loc='best')

plt.subplot(224)
plt.plot(x_np,y_softplus,label='softplus')
plt.ylim((-0.2,6))
plt.legend(loc='best')

plt.show()
```
